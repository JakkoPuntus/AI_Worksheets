```python
#1.3.1
import numpy as np

arr = np.zeros((8, 8)) #создание матрицы из нулей
arr[::2, ::2] = 1 #чётные элементы чётных строк равны 1
arr[1::2, 1::2] = 1 #нечётные элементы нечётных строк равны 1

print(arr)
```

```python
#1.3.2

arr =  np.zeros((5, 1)) + np.arange(5)
print(arr)
```

```python
#1.3.3

arr = np.random.random((3, 3, 3))

print(arr)
```

```python
#1.3.4
size = (3, 3)

arr = np.zeros(size)

arr[0, :] = 1  # Top row
arr[-1, :] = 1  # Bottom row
arr[:, 0] = 1  # Left column
arr[:, -1] = 1  # Right column

print(arr)
```

```python
#1.3.5
arr = np.random.random((10))
arr.sort()
arr = arr[::-1]

print(arr)
```

```python
#2.2.1
import pandas as pd

#Создание двух точек a и b
a = pd.Series([1, 2, 3])
b = pd.Series([1, 1, 3])

#Вычисление евклидова расстояния
distance = np.sqrt(np.sum((a - b) ** 2))

print(distance)
```

```python
#3.3.2
from sklearn import preprocessing

url = 'https://raw.githubusercontent.com/akmand/datasets/master/iris.csv'
data = pd.read_csv(url)

minmax_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
z_scaler = preprocessing.StandardScaler()

standardized_first = minmax_scaler.fit_transform(data[['sepal_length_cm']])
standardized_second = z_scaler.fit_transform(data[['sepal_width_cm']])

print('First', standardized_first)
print('Second', standardized_second)

```
