```python
#4.3.2
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return np.sqrt(1 + np.e**x**0.5 + np.cos(x**2))/(abs(1-np.sin(x)**3))

x = np.arange(0, 10, 1)
y = np.abs(f(x))

plt.plot(x, y)
plt.scatter(x[:len(x)//2], y[:len(y)//2])
plt.show()
```

```python
#4.3.5
import numpy as np

def calculator(x, y, operator):
    if operator == '+':
        result = x + y
    elif operator == '-':
        result = x - y
    elif operator == '*':
        result = x * y
    elif operator == '/':
        result = x / y
    elif operator == 'exp':
        result = np.exp(x + y)
    elif operator == 'sin':
        result = np.sin(x + y)
    elif operator == 'cos':
        result = np.cos(x + y)
    elif operator == '^':
        result = np.power(x, y)
    else:
        result = 'Неверный оператор'
    return result

# Пример использования калькулятора
x = 2
y = 3
operator = '+'
result = calculator(x, y, operator)
print(result)
```
